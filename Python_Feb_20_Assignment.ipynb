#### Q1. Explain GET and POST methods.

```python
'''
GET Method:
The GET method is used to retrieve data from the server. When a client sends a GET request, it includes parameters in 
the URL itself. The parameters are typically encoded in the query string of the URL. For example, the following URL contains a 
query string with two parameters, "name" and "age":

http://example.com/page?name=John&age=25

Since the parameters are included in the URL, the GET method is limited to the maximum length of a URL, which is typically 
around 2000 characters. The GET method is idempotent, which means that multiple identical requests will have the same effect as 
a single request. In other words, sending the same GET request multiple times will not have any side effects.

POST Method:
The POST method is used to submit data to be processed by the server. Unlike GET requests, POST requests include parameters in 
the request body, rather than in the URL. This allows for larger amounts of data to be sent to the server. The POST method is 
not idempotent, which means that sending the same request multiple times may have different effects each time. This is because 
the server may perform a different action based on the data included in the request body

'''
```

#### Q2. Why is request used in Flask? 

```python
'''
In Flask, the request module is used to access incoming request data such as form data, query parameters, and HTTP headers. It 
provides a convenient way to handle and process data submitted by clients (such as web browsers) to the server.

Here are some of the main reasons why the request module is used in Flask:

Accessing form data: When a user submits a form on a web page, the data from the form is sent to the server in the request 
object. The request module provides a simple way to access this data and extract the values submitted by the user.

Query parameters: The request module also allows you to access query parameters that are included in the URL of the request. 
This is useful when building APIs or web applications that accept parameters in the URL.

HTTP headers: The request module can be used to access HTTP headers that are sent by the client along with the request. This is 
useful when building APIs or web applications that need to authenticate clients or check for specific headers.

File uploads: Flask also provides functionality for uploading files, and the request module can be used to handle these file 
uploads and extract the uploaded file data.

'''
```

#### Q3. Why is redirect() used in Flask? 

```python
'''
In Flask, the redirect() function is used to redirect the user to a different URL. This can be useful in a number of scenarios, 
such as after a user has submitted a form or completed an action on a web page. Here are some of the main reasons why redirect() 
is used in Flask:

Redirect after form submission: When a user submits a form on a web page, it is often desirable to redirect them to a different 
page to display the results of their submission or to show them a confirmation message. The redirect() function can be used to 
send the user to a different URL after the form has been submitted.

Redirect after authentication: When building web applications that require authentication, it is common to redirect the user to 
a different page after they have successfully logged in or registered. The redirect() function can be used to send the user to 
a different page based on their authentication status.

Redirect after error handling: When handling errors in Flask, it may be desirable to redirect the user to a different page to 
display an error message or to provide them with further instructions. The redirect() function can be used to send the user to 
a different page after an error has been encountered.

SEO and URL structure: Redirects can also be useful for search engine optimization (SEO) and for maintaining a consistent URL 
structure on a website. For example, if a URL changes but you want to maintain the original URL for SEO purposes, you can use a 
redirect to send users to the new URL while still maintaining the old URL.
'''
```

#### Q4. What are templates in Flask? Why is the render_template() function used?

```python
'''
In Flask, templates are used to generate HTML pages dynamically. A template is a file that contains a mixture of HTML code and 
placeholders for dynamic content that is generated by Flask when the page is requested. The Flask framework uses the Jinja2 
templating engine to process templates.

Here are some of the main reasons why templates are used in Flask:

Dynamic content: Templates are used to generate HTML pages that include dynamic content, such as data from a database, user 
input, or content from an external API.

Separation of concerns: Templates allow for a separation of concerns between the logic that generates the dynamic content and 
the presentation of that content. This makes it easier to maintain and update the website or web application over time.

Reusability: Templates can be reused across multiple pages of a website or web application, making it easier to maintain 
consistency in the design and layout of the site.

The render_template() function in Flask is used to render a template and generate an HTML page. This function takes the name of 
the template file as an argument, along with any variables that should be passed to the template for rendering.
'''
```

#### Q5. Create a simple API. Use Postman to test it. Attach the screenshot of the output in the Jupyter Notebook.

```python
from flask import Flask,request ,render_template , jsonify

app = Flask(__name__)

@app.route('/postman_action',methods=['POST'])
def math_ops1():
    if(request.method == 'POST'):
        ops = request.json['operation']
        num1 = int(request.json['num1'])
        num2 = int(request.json['num2'])
        if ops == 'add':
            r = num1+num2
            result = "The sum of " + str(num1) + ' and ' + str(num2) + " is " + str(r)
        if ops == 'subtract':
            r = num1-num2
            result = "The subtract of " + str(num1) + ' and ' + str(num2) + " is " + str(r)
        if ops == 'multiply':
            r = num1*num2
            result = "The multiply of " + str(num1) + ' and ' + str(num2) + " is " + str(r)
        if ops == 'divide':
            r = num1/num2
            result = "The divide of " + str(num1) + ' and ' + str(num2) + " is " + str(r)
            
        return jsonify(result) 
    
if __name__=="__main__":
    app.run(host="0.0.0.0", port = 5003)
```
