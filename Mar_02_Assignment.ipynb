#### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

```python
'''
A scatter plot is a type of plot that displays the relationship between two variables by showing the distribution of points in a
two-dimensional space. Each point represents an observation in the data, and the position of the point on the x-y plane 
corresponds to the values of the two variables for that observation.
'''

import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x,y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Scatter Plot x vs y")
```

#### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data: 

```python
'''
The subplot() function is used to create multiple plots in a single figure. It allows us to arrange multiple plots in a grid 
format, with a specified number of rows and columns, and plot data in each of the subplots.
'''

import numpy as np
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
    
fig, axs = plt.subplots(2, 2)

# plot the data
axs[0, 0].plot(x1, y1)
axs[0, 1].plot(x2, y2)
axs[1, 0].plot(x3, y3)
axs[1, 1].plot(x4, y4)

# set the titles
axs[0, 0].set_title('Line 1')
axs[0, 1].set_title('Line 2')
axs[1, 0].set_title('Line 3')
axs[1, 1].set_title('Line 4')

# set the x and y labels
fig.text(0.5, 0.04, 'X-axis', ha='center')
fig.text(0.04, 0.5, 'Y-axis', va='center', rotation='vertical')

# adjust the spacing between subplots
plt.subplots_adjust(hspace=0.4)

plt.show()
```

#### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

```python
'''
A bar plot is a type of plot that displays categorical data with rectangular bars, where the length or height of each bar 
corresponds to the value of the category being plotted. Bar plots are often used to compare data across different categories.
'''
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
plt.xlabel("Companies")
plt.ylabel("Profit")
plt.title("Companies vs Profit Bar Chart")
```

#### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

```python
'''
A box plot is a type of plot that displays the distribution of a dataset by showing the median, quartiles, and outliers of the 
data. It consists of a box (hence the name) that spans the interquartile range (IQR) of the data, with a line inside the box 
indicating the median. Whiskers extend from the top and bottom of the box to the smallest and largest data points within 1.5 
times the IQR of the box. Points outside of this range are considered outliers and are plotted individually.

Box plots are useful for visualizing the spread and skewness of a dataset, as well as identifying outliers.
'''

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data = [box1, box2]

plt.boxplot(data)
plt.title("Box Plot")
plt.xlabel("Box")
plt.ylabel("Value")
```
