#### Q1. How can you create a Bokeh plot using Python code?

```python
import bokeh.io
bokeh.io.output_notebook()
from bokeh.plotting import figure, output_file, show
output_file('plot.html')
x = [2,3,4,5,6]
y = [4,5,6,7,8]
p = figure(title="My Bokeh Plot", x_axis_label='X Axis', y_axis_label='Y Axis')
p.circle(x,y)
show(p)
```

#### Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

```python
from bokeh.models import ColumnDataSource
output_file('glyphs.html')
x = [1,2,3,4,5]
y = [6,7,4,2,5]
source = ColumnDataSource(data = dict(x = x, y = y))
p = figure(title='Glyphs Example', plot_width=400, plot_height=400)
p.circle('x', 'y', size=10, source=source, color='red', alpha=0.5)
show(p)
```

#### Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?

```python
from bokeh.sampledata.penguins import data
output_file('data.html')
p = figure(title = "Penguins Data")
p.xaxis.axis_label = 'bill_length_mm'
p.yaxis.axis_label = 'bill_depth_mm'
p.circle(data['bill_length_mm'], data['bill_depth_mm'], legend_label = "length vs depth")
show(p)
```

#### Q5. How can you embed a Bokeh plot into a web page or dashboard using Flask or Django?

```python
'''
Bokeh is a powerful data visualization library that can be used to create interactive plots and dashboards. Flask and Django 
are popular web frameworks for building web applications in Python. Embedding a Bokeh plot into a web page or dashboard using 
Flask or Django involves creating a Bokeh plot and then integrating it into your web application using either framework's 
templating system.

To embed a Bokeh plot into a web page or dashboard using Flask, you can follow these steps:

1.Create your Bokeh plot using the Bokeh library in a separate Python script.
2. Define a Flask route that will render your web page or dashboard.
3. Inside the route function, use the Bokeh components() function to generate the JavaScript and HTML code required to embed your 
Bokeh plot into the web page or dashboard.
4. Pass the JavaScript and HTML code to the Flask template as context variables.
5. Render the Flask template with the context variables.

'''
from flask import Flask, render_template
from bokeh.plotting import figure
from bokeh.embed import components

app = Flask(__name__)

@app.route('/')
def index():
    # Create a Bokeh plot
    plot = figure()
    plot.line([1, 2, 3], [4, 5, 6])

    # Generate the JavaScript and HTML code required to embed the plot into the web page
    script, div = components(plot)

    # Pass the JavaScript and HTML code to the Flask template as context variables
    return render_template('index.html', plot_script=script, plot_div=div)

if __name__ == '__main__':
    app.run(debug=True)
```
